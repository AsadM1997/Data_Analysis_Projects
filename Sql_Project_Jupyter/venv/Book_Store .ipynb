```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS Orders (
    Order_ID INT AUTO_INCREMENT PRIMARY KEY,
    Customer_ID INT,
    Book_ID INT,
    OrderDate DATE,
    Quantity INT,
    Total_Amount DECIMAL(10,2),
    FOREIGN KEY (Customer_ID) REFERENCES Customers(Customer_ID),
    FOREIGN KEY (Book_ID) REFERENCES Books(Book_ID)
) ENGINE=InnoDB;
""")
conn.commit()
```

```python
# --9) Calculate the stock remaining after fulfilling all orders:
df = pd.read_sql_query("""
SELECT 
    b.Book_ID,
    b.Title,
    b.Stock - IFNULL(SUM(o.Quantity), 0) AS Stock_Remaining
FROM Books b
LEFT JOIN Orders o 
    ON b.Book_ID = o.Book_ID
GROUP BY b.Book_ID, b.Title, b.Stock
ORDER BY Stock_Remaining ASC
""", conn)

df.head()

# 🧠 Explanation

# LEFT JOIN → includes all books, even those with no orders.

# SUM(o.Quantity) → total number of copies sold for that book.

# IFNULL(..., 0) → replaces NULL with 0 if a book has no orders.

# b.Stock - SUM(o.Quantity) → remaining stock after orders.

# ORDER BY Stock_Remaining ASC → books with lowest stock first.
```
